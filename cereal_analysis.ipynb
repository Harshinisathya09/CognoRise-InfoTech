## Relationship between various manufacturers and their ratings ##

```python
df_sorted = df.sort_values(by='rating', ascending=False)
plt.figure(figsize=(8, 5))
sns.barplot(data=df_sorted,x='mfr',y='rating',errorbar=None)
plt.title('Manufacturers and their ratings')
plt.xlabel('Manufacturers')
plt.ylabel('Ratings')
```

## Relationship between Calories content and ratings ##

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=df,x='calories',y='rating')
plt.title('Calories and ratings')
```

***It is evident that low calorie cereals have the highest rating***

## Relationship between manufacturers and type of cereal ##

```python
plt.figure(figsize=(8,5))
sns.histplot(binwidth=1,
            x='mfr',
            hue='type',
            data=df,
            stat="count",
            multiple="dodge",
            kde=True)
plt.title('Type of cereal by each manufacturer')
plt.xlabel('Manufacturer')
```

***It is observed that most of the manufacturers produce COLD type cereals***

## Relationship between sugar content and ratings ##

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=df,x='sugars',y='rating')
plt.title('Sugar and ratings')
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x='type',y='calories',data=df,errorbar=None)
plt.title('Type Vs Calories')
plt.xlabel('Type')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
num=['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups','rating']
sns.heatmap(df[num].corr(), annot=True, cmap='coolwarm',ax=ax)
plt.show()
```

## Potassium and Fiber has the highest correlation ##

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=df,x='potass',y='fiber')
plt.title('Potassium Vs Fiber')
plt.xlabel('Potassium')
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=df,x='potass',y='rating')
plt.title('Potassium and ratings')
plt.figure(figsize=(8, 6))
sns.regplot(data=df,x='fiber',y='rating')
plt.title('Fiber and ratings')
```
