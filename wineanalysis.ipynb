```python
plt.figure(figsize=(10,6))
plt.title("Data distribution")
sns.histplot(data=wine)
```

```python
sns.histplot(x='fixed acidity',data=wine,kde=True)
plt.title("Distribution of fixed acidity")
plt.xlabel("Fixed acidity")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(wine.corr(),annot=True,linewidth=0.5,center=0,cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(x='quality',y='alcohol',data=wine,palette='GnBu_d')
```

```python
sns.boxplot(x='quality',y='pH',data=wine,palette='GnBu_d')
plt.title("Boxplot of Quality and pH")
plt.show()
```
