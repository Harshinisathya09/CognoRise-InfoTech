# Employment Salary Analysis #

# Importing packages #

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Descriptive analysis ###

```python
df=pd.read_csv('ds_salaries.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

### Distribution of Working years ###

```python
plt.figure(figsize=(10,6))
sns.histplot(x='work_year',data=df,kde=True,color='orange')
plt.title('Working year distribution')
plt.xlabel('Work Year')
```

### Distribution of experience level ###

```python
plt.figure(figsize=(10,6))
sns.histplot(x='experience_level',data=df,kde=True,color='blue')
plt.title('Distribution of experience level')
plt.xlabel('Experience level')
```

### Distribution of employment type ###

```python
plt.figure(figsize=(10,6))
sns.histplot(x='employment_type',data=df,color='violet')
plt.title('Distribution of employment type')
plt.xlabel('Employment type')
```

### Distribution of Job title ###

```python
plt.figure(figsize=(10,6))
ax=sns.histplot(x='job_title',data=df,kde=True,color='green')
plt.xticks(rotation=90)
plt.xlabel('Job title')
ax.set_title('Distribution of Job title')
```

### Distribution of type of currency ###

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='salary_currency',kde=True,color='black')
plt.title('Distribution of type of currency')
plt.xlabel('Type of currency')
```

### Distribution of salary in USD ###

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='salary_in_usd',kde=True,color='red')
plt.title('Distribution of salary in usd')
plt.xlabel('Salary in USD')
```

### Distribution of employee residence ###

```python
plt.figure(figsize=(10,8))
ax=sns.histplot(data=df,x='employee_residence',kde=True)
plt.xticks(rotation=90)
plt.xlabel('Employee Residence')
ax.set_title('Distribution of employee residence')
```

### Distribution of Company's location ###

```python
plt.figure(figsize=(11,8))
ax=sns.histplot(data=df,x='company_location',kde=True)
plt.xlabel("Company's location")
plt.xticks(rotation=90)
ax.set_title('Distribution of company location')
```

### Distribution of company size ###

```python
plt.figure(figsize=(10,6))
sns.histplot(x='company_size',data=df,kde=True)
plt.xlabel('Company size')
plt.title('Distribution of company size')
```

## Relationship between Job title and Salary ##

```python
plt.figure(figsize=(8, 5))
ax=sns.barplot(data=df,x='job_title',y='salary',errorbar=None)
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.xticks(rotation=90)
ax.set_title('Job titles Vs Salary')
```

```python
plt.figure(figsize=(8, 5))
ax=sns.barplot(data=df,x='job_title',y='salary_in_usd',errorbar=None)
plt.xlabel('Job title')
plt.ylabel('Salary in USD')
plt.xticks(rotation=90)
ax.set_title('Job titles Vs Salary in USD')
```

## Relationship between company size and employment type ##

```python
plt.figure(figsize=(8,5))
sns.histplot(binwidth=1,
            x='company_size',
            hue='employment_type',
            data=df,
            stat="count",
            multiple="dodge")
plt.title('Company size Vs employment type')
plt.xlabel('Company size')
```

## Relationship between Job title and remote ratio ##

```python
plt.figure(figsize=(8, 5))
ax=sns.barplot(data=df,x='job_title',y='remote_ratio',errorbar=None)
plt.xlabel('Job title')
plt.ylabel('Remote ratio')
plt.xticks(rotation=90)
ax.set_title('Job titles Vs remote ratio')
```

## Relationship between Job title and employment type ##

```python
plt.figure(figsize=(10,8))
ax=sns.histplot(binwidth=1,
            x='job_title',
            hue='employment_type',
            data=df,
            stat="count",
            multiple="dodge")
plt.xlabel('Job title')
plt.xticks(rotation=90)
ax.set_title('Job Vs Type')
```

## Relationship between employment type and salary ##

```python
plt.figure(figsize=(8, 5))
ax=sns.barplot(data=df,x='employment_type',y='salary',errorbar=None)
ax.set_title('Employment type Vs Salary')
plt.xlabel('Employement type')
plt.ylabel('Salary')
```

## Relationship between experience level and salary ##

```python
plt.figure(figsize=(8, 5))
ax=sns.barplot(data=df,x='experience_level',y='salary',errorbar=None)
ax.set_title('Experience level Vs Salary')
plt.xlabel('Experience level')
plt.ylabel('Salary')
```
